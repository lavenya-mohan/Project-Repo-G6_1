```python
## SLIDE1 
import matplotlib.pyplot as plt

west_coast_population = {
    "Los Angeles": 3980400,
    "Seattle": 753675,
    "Portland": 652503
}

east_coast_population = {
    "New York City": 8419600,
    "Jacksonville": 949611,
    "Philadelphia": 1584200
}

combined_population = {**west_coast_population, **east_coast_population}
labels = list(combined_population.keys())
sizes = list(combined_population.values())
colors = ["#8B4513", "#5C4033", "#D2691E", "#1E90FF", "#4682B4", "#5F9EA0"]

fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
ax.set_title("Population Distribution Across East and West Coast Major Cities")

table_data = [[city, f"{pop:,}"] for city, pop in combined_population.items()]
table = plt.table(cellText=table_data, colLabels=["City", "Population"], cellLoc="center", loc="bottom", bbox=[-0.3, -0.5, 1.6, 0.3])
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 1.2)  # Adjust table size

plt.subplots_adjust(left=0.2, bottom=0.4)  # Adjust layout to make room for the table
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Data for coffee quality attributes and scores by coast
data = {
    'Attribute': ['Flavor', 'Aroma', 'Acidity', 'Body', 'Balance'],
    'West Coast': [8.2, 8.0, 7.8, 8.3, 8.1],
    'East Coast': [7.9, 7.5, 8.1, 7.6, 7.7]
} 

# Create a DataFrame
df = pd.DataFrame(data) 

# Set up radar chart categories and values
categories = list(df['Attribute'])
values_west = df['West Coast'].values
values_east = df['East Coast'].values

# Calculate angles for radar chart
num_vars = len(categories)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Complete the loop for radar chart by appending the first value at the end
values_west = np.append(values_west, values_west[0])
values_east = np.append(values_east, values_east[0])
angles += angles[:1]

# Create radar chart
fig, ax = plt.subplots(figsize=(8, 10), subplot_kw=dict(polar=True))

ax.plot(angles, values_west, color="brown", linewidth=2, label="West Coast")
ax.fill(angles, values_west, color="brown", alpha=0.25)

ax.plot(angles, values_east, color="darkblue", linewidth=2, label="East Coast")
ax.fill(angles, values_east, color="darkblue", alpha=0.25)

# Set category labels on the axes
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

# Title and legend
ax.legend(loc="upper right", bbox_to_anchor=(1.1, 1.1))

# Add a table below the radar chart
table_data = df.values
table = plt.table(cellText=table_data, colLabels=df.columns, cellLoc='center', loc='bottom', bbox=[-0.2, -0.5, 1.4, 0.3])
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 1.5)

# Adjust layout to fit table below the chart
plt.subplots_adjust(left=0.2, bottom=0.4)
plt.show()
```
